```python
import torch
import torch.nn.functional as F
```

```python
words = open('names.txt','r').read().splitlines()
```

```python
char = sorted(list(set(''.join(words))))
```

```python
stoi = {s:i+1 for i,s in enumerate(char)}
stoi['.']=0
```

```python
isto = {i:s for s,i in stoi.items()}
```

```python
N = torch.zeros([27,27,27],dtype=torch.int32)
```

```python
N
```

```python
for w in words:
  chs = ['.'] + list(w) + ['.']
  for ch1, ch2, ch3 in zip(chs, chs[1:], chs[2:]):
    ix1 = stoi[ch1]
    ix2 = stoi[ch2]
    ix3 = stoi[ch3]
    N[ix1, ix2, ix3] += 1
```

```python
N
```

```python
g = torch.Generator().manual_seed(2147483647)
```

```python
P = (N + 1).float()
P /= P.sum(2, keepdim=True)
```

```python
for i in range(10):
  ix1 = 0
  ix2 = 0
  out =[]
  while True:
    p = P[ix1, ix2]
    ix1 = ix2
    ix = torch.multinomial(p, num_samples=1, replacement=True, generator=g).item()
    out.append(isto[ix])
    if ix == 0:
      break
    ix1, ix2 = ix2, ix
  print(''.join(out))
```

```python
for w in words:
  n = 0
  likelihood = 0.0
  chs = ['.'] + list(w) + ['.']
  for ch1, ch2, ch3 in zip(chs, chs[1:], chs[2:]):
    ix1 = stoi[ch1]
    ix2 = stoi[ch2]
    ix3 = stoi[ch3]
    prob = P[ix1, ix2, ix3]
    logprob = torch.log(prob)
    likelihood += logprob
    n += 1

    print(f'{ch1}{ch2}{ch3}: {prob:.4f} {logprob:.4f} {likelihood:.4f}')

print(f"{likelihood=}")
nll = -likelihood
print(f'{nll/n=}')
```

```python
xs = []
ys = []
for w in words:
  chs = ['.'] + list(w) + ['.']
  for ch1, ch2, ch3 in zip(chs, chs[1:], chs[2:]):
    xs.append((stoi[ch1],stoi[ch2]))
    ys.append(stoi[ch3])
xs = torch.tensor(xs)
ys = torch.tensor(ys)
```

```python
W = torch.randn((27*2, 27), requires_grad=True)
```

```python
for i in range(10):
  x1 = F.one_hot(xs[:,0]).float()
  x2 = F.one_hot(xs[:,1]).float()
  xenc = torch.cat([x1,x2],dim =1 )
  logits = xenc @ W
  probs = F.log_softmax(logits, dim=1)
  loss = -probs[torch.arange(xs.shape[0]), ys].mean()
  W.grad = None
  loss.backward()
  W.data += -0.4 * W.grad
  print(loss.item())
```

```python
for i in range(100):
  x1 = F.one_hot(xs[:,0]).float()
  x2 = F.one_hot(xs[:,1]).float()
  xenc = torch.cat([x1,x2],dim =1 )
  logits = xenc @ W
  probs = F.log_softmax(logits, dim=1)
  loss = -probs[torch.arange(xs.shape[0]), ys].mean()
  W.grad = None
  loss.backward()
  W.data += -0.6 * W.grad
  print(loss.item())
```

```python
for i in range(10):
  out = []
  ix1 = 0
  ix2 = 0
  while True:
    t1 = F.one_hot(torch.tensor([ix1]),num_classes=27).float()
    t2 = F.one_hot(torch.tensor([ix2]), num_classes=27).float()
    xenc = torch.cat([t1,t2],dim =1)
    logits = xenc @ W
    probs = F.softmax(logits, dim=1)
    ix = torch.multinomial(probs, num_samples=1, replacement=True, generator=g).item()
    out.append(isto[ix])
    if ix == 0:
      break
    ix1 , ix2 = ix2, ix
  print(''.join(out))
```
